# Intro to Probability Lab

## Challenge 1 — Ironhack Airlines (Overbooking)

- Plane has **450 seats**
- Each passenger has a **3% chance of missing** the flight → shows up with probability **p = 0.97**
- The airline sells **460 tickets**

**Question:** what is the chance they have a seat for every passenger who shows up?

Let `X` = number of passengers who show up. Each of the 460 ticket-holders shows up independently with probability 0.97, so `X ~ Binomial(n=460, p=0.97)`. Everyone gets a seat as long as `X <= 450`, so the answer is `P(X <= 450)` = the binomial CDF at 450.

```python
from scipy.stats import binom

n, p, seats = 460, 0.97, 450   # 460 tickets, 97% show-up rate, 450 seats
prob = binom.cdf(seats, n, p)  # P(X <= 450)

print(f"P(seats for all passengers) = {prob:.4f}  ({prob*100:.2f}%)")
```

### Findings

`P(X <= 450) ≈ 0.8845`. There is about an **88.45%** chance that every passenger who shows up has a seat — meaning roughly a **12%** chance the airline oversells and has to bump at least one passenger.

## Challenge 2 — Ironhack Call Center

- Probability of resolving a complaint on any single attempt: **p = 0.3**

**Question:** what is the probability the representative needs **at least three attempts** before resolving a complaint?

Let `X` = the attempt number on which the first success occurs, so `X ~ Geometric(p=0.3)`. "Needs at least three attempts" means the first success happens on attempt 3 or later (`X >= 3`), which is exactly the case where the first two attempts both fail:

`P(X >= 3) = (1 - p)^2 = 0.7^2 = 0.49`

```python
from scipy.stats import geom

p = 0.3
prob = geom.sf(2, p)   # P(X > 2) = P(X >= 3)

print(f"Manual (1-p)^2 = {(1-p)**2:.4f}")
print(f"P(at least 3 attempts) = {prob:.4f}  ({prob*100:.2f}%)")
```

### Findings

`P(X >= 3) = 0.49`. There is a **49%** chance the representative needs at least three attempts — i.e., the first two attempts both fail before a resolution.

## Challenge 3 — Ironhack Website (Server Load)

- Average traffic: **500 visits per hour**
- Server capacity: handles up to **550 visits per hour**

**Question:** what is the probability the server is overwhelmed?

Counts of events over a fixed interval follow a **Poisson** distribution. Let `X` = visits in one hour, so `X ~ Poisson(λ=500)`. The server is overwhelmed when traffic exceeds capacity (`X > 550`):

`P(X > 550) = 1 - P(X <= 550)`

```python
from scipy.stats import poisson

lam, capacity = 500, 550
prob = poisson.sf(capacity, lam)   # P(X > 550) = 1 - P(X <= 550)

print(f"P(server overwhelmed) = {prob:.4f}  ({prob*100:.2f}%)")
```

### Findings

`P(X > 550) ≈ 0.0129`. There is only about a **1.29%** chance in any given hour that traffic exceeds the server's capacity of 550 visits — so the server is rarely overwhelmed.

### Over a full day (24 hours)

Treat each hour as an independent trial with overwhelm probability `q = P(X > 550) ≈ 0.0129`. The server is overwhelmed **at least once** during the day unless it survives all 24 hours:

`P(at least once in 24h) = 1 - (1 - q)^24`

```python
q = poisson.sf(550, 500)      # hourly P(overwhelmed)
day = 1 - (1 - q) ** 24       # at least once during 24 hours

print(f"Hourly q          = {q:.4f}")
print(f"P(>=1 in 24 hours) = {day:.4f}  ({day*100:.2f}%)")
```

### Findings

`P(>=1 in 24h) ≈ 0.2677`. Although any single hour has only a ~1.3% chance of overload, accumulated across 24 hours there is roughly a **26.77%** chance the server is overwhelmed at some point during the day.

## Challenge 4 — Ironhack Helpdesk (Time Between Arrivals)

- On average, a customer arrives every **10 minutes**

**Question:** what is the probability the next customer arrives within the next **5 minutes**?

The time between Poisson arrivals follows an **Exponential** distribution. With a mean of 10 minutes, the rate is `λ = 1/10` per minute. Let `T` = waiting time, `T ~ Exponential(λ=0.1)`:

`P(T <= 5) = 1 - e^(-λ·5) = 1 - e^(-0.5)`

```python
from scipy.stats import expon

mean = 10                          # average minutes between arrivals
prob = expon.cdf(5, scale=mean)    # P(T <= 5); scale = mean = 1/lambda

print(f"P(next arrival within 5 min) = {prob:.4f}  ({prob*100:.2f}%)")
```

### Findings

`P(T <= 5) ≈ 0.3935`. There is about a **39.35%** chance the next customer arrives within the next 5 minutes.

### Break time

An employee can take a 5-minute break if **no customer arrives for 15 minutes** — i.e. the waiting time exceeds 15 minutes:

`P(T > 15) = e^(-λ·15) = e^(-1.5)`

```python
prob_break = expon.sf(15, scale=10)   # P(T > 15)

print(f"P(employee takes a break) = {prob_break:.4f}  ({prob_break*100:.2f}%)")
```

### Findings

`P(T > 15) ≈ 0.2231`. There is about a **22.31%** chance that no customer arrives for 15 minutes, allowing the employee to take a break.

## Challenge 5 — Bird Weights (Normal Distribution)

- Weights follow a **Normal** distribution: mean **μ = 150 g**, standard deviation **σ = 10 g**

**Question:** for a randomly selected bird, what is the probability its weight is between **140 and 160 grams**?

Let `X ~ Normal(μ=150, σ=10)`. We want `P(140 <= X <= 160) = P(X <= 160) - P(X <= 140)`. Note 140 and 160 are exactly μ ± 1σ, so this is the classic ~68% within one standard deviation.

```python
from scipy.stats import norm

mu, sigma = 150, 10
prob = norm.cdf(160, mu, sigma) - norm.cdf(140, mu, sigma)

print(f"P(140 <= X <= 160) = {prob:.4f}  ({prob*100:.2f}%)")
```

### Findings

`P(140 <= X <= 160) ≈ 0.6827`. There is about a **68.27%** chance a randomly selected bird weighs between 140 and 160 grams — consistent with the empirical rule that ~68% of a normal distribution lies within one standard deviation of the mean.

## Challenge 6 — Electronic Component Lifetime (Exponential)

- Lifetime follows an **Exponential** distribution with a mean of **50 hours**

**Question:** what is the probability the component fails within the first **30 hours**?

Let `T` = lifetime, `T ~ Exponential` with mean 50 (rate `λ = 1/50`). We want `P(T <= 30)`:

`P(T <= 30) = 1 - e^(-30/50) = 1 - e^(-0.6)`

```python
mean_life = 50                          # mean lifetime in hours
prob = expon.cdf(30, scale=mean_life)   # P(T <= 30)

print(f"P(fails within 30 hours) = {prob:.4f}  ({prob*100:.2f}%)")
```

### Findings

`P(T <= 30) ≈ 0.4512`. There is about a **45.12%** chance the component fails within the first 30 hours.
